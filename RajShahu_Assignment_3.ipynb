```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
```

```python
# Load the Titanic dataset
train_data = pd.read_csv('train.csv')
train_data.head()
```


### Data Preprocessing

1. Handle missing values in the Age column.
2. Encode the Sex column to convert it from categorical to numerical.
3. Split the data into training (first 200 samples) and testing sets (samples 201 to 240).


```python
# Handle missing values in the Age column by filling with mean
train_data['Age'].fillna(train_data['Age'].mean(), inplace=True)

# Encode the Sex column
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
train_data['Sex_encoded'] = encoder.fit_transform(train_data['Sex'])

# Split the data into training and testing sets
train_subset = train_data.iloc[:200]
test_subset = train_data.iloc[200:240]

# Extract relevant columns
X_train_age = train_subset[['Age']]
X_train_gender = train_subset[['Sex_encoded']]
X_train_both = train_subset[['Age', 'Sex_encoded']]
y_train = train_subset['Survived']

X_test_age = test_subset[['Age']]
X_test_gender = test_subset[['Sex_encoded']]
X_test_both = test_subset[['Age', 'Sex_encoded']]
y_test = test_subset['Survived']

X_train_age.shape, X_train_gender.shape, X_train_both.shape
```


### 1) Model Construction

#### **Add a part to the program so that both age and gender are considered as the input variables (features). Construct the logistic regression model and estimate the parameters of the log odd function using the first 200 data points. Print the parameters of the model.**


```python
# Model using only Age as a feature
model_age = LogisticRegression()
model_age.fit(X_train_age, y_train)

# Model using only Gender as a feature
model_gender = LogisticRegression()
model_gender.fit(X_train_gender, y_train)

# Model using both Age and Gender as features
model_both = LogisticRegression()
model_both.fit(X_train_both, y_train)
print("\n====================================================================\n")
print("Parameters for model with 'Age' as feature:")
print("Intercept:", model_age.intercept_)
print("Coefficients:", model_age.coef_)
print("\n====================================================================")

print("\nParameters for model with 'Gender' as feature:")
print("Intercept:", model_gender.intercept_)
print("Coefficients:", model_gender.coef_)

print("\n====================================================================")
print("\nParameters for model with both 'Age' and 'Gender' as features:")
print("Intercept:", model_both.intercept_)
print("Coefficients:", model_both.coef_)
print("\n====================================================================")
```


### 2 Model Evaluation

#### **Calculate the value of the log-loss function in this case for the sample points 201 to 240. Print the values of the log-loss function (average of the penalty for these 40 samples) for the three models constructed so far, that is, the model consider only age as feature, the model considers only gender as feature, and the model considers both of them as features.**


```python
# Predict probabilities for each model
y_pred_prob_age = model_age.predict_proba(X_test_age)
y_pred_prob_gender = model_gender.predict_proba(X_test_gender)
y_pred_prob_both = model_both.predict_proba(X_test_both)

# Compute log-loss for each model
log_loss_age = log_loss(y_test, y_pred_prob_age)
log_loss_gender = log_loss(y_test, y_pred_prob_gender)
log_loss_both = log_loss(y_test, y_pred_prob_both)

print("\nLog-Loss for model with 'Age' as feature:", log_loss_age)
print("\n===============================================================================\n")
print("Log-Loss for model with 'Gender' as feature:", log_loss_gender)
print("\n===============================================================================\n")
print("Log-Loss for model with both 'Age' and 'Gender' as features:", log_loss_both)
```


### 3) Conclusion

####**Which one is the most accurate one? (Add you answer as a text box to the program).**




We know that, lower values of log-loss indicate better performance.

- The model considering only the Age feature has the highest log-loss, making it the least accurate.
- The model considering both Age and Gender as features has the lowest log-loss, making it the most accurate.
- The model considering only the Gender feature also performs well, with a log-loss value close to the combined model.

**Hence, the most accurate model is the one that considers both Age and Gender as features.**
